```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import cufflinks as cf
```

```
df= sns.load_dataset('tips')
df
```

```
df['size'].unique()
```

# For Quantitative Data

# Distribution Plot (Histogram)

```
sns.histplot(df['total_bill'] , bins = 20 , kde = True)
```

```
plt.subplot(1,2,1)
sns.histplot(df['total_bill'])

plt.subplot(1,2,2)
sns.histplot(df['tip'])
```

#Joint Plot

```
sns.jointplot(x = 'total_bill' , y= 'tip', data = df , kind = 'scatter')
```

```
sns.jointplot(x = 'total_bill' , y= 'tip', data = df , kind = 'kde')
```

```
sns.jointplot(x = 'total_bill' , y= 'tip', data = df , kind = 'reg')
```

# Pair Plot

```
sns.pairplot(df , hue= 'size' , palette = 'rainbow')
```

# Categorical plots

```
sns.countplot(x = df['sex'] , hue = df['smoker'])
```

# Bar Plot

```
sns.barplot(x = df['sex'] , y= df['total_bill'] )
```

# Box plot

```
sns.boxplot(x = 'tip' , y = 'day' , data = df , palette = 'rainbow')
```

# Violin plot

```
sns.violinplot(x= 'day',y= 'total_bill' , data = df , palette= 'rainbow')
```

#Strip Plot

```
sns.stripplot(x= 'day',y= 'total_bill' , data = df)
```

# Swarm Plot

```
sns.swarmplot(x= 'day',y= 'total_bill' , data = df)
```

# Swarm Plot and Violin Plot

```

sns.violinplot(x= 'tip',y= 'day' , data = df)
sns.swarmplot(x= 'tip',y= 'day' , data = df)
```

# Matrix Plots

```
ds = sns.load_dataset('flights')
ds
```

```
tipsCorr = df[['total_bill' , 'tip', 'size']]
tipsCorr
```

```
tipsCorr.corr()
```

```
sns.heatmap(tipsCorr.corr() , annot = True)
```

# Cluster Map

```
sns.clustermap(tipsCorr.corr())
```

# Pivot Heatmap

```
pvtFlights = ds.pivot_table(values = 'passengers' , columns = 'year', index ='month')
```

```
pvtFlights
```

```
sns.heatmap(pvtFlights)
```

# Regression Plot

```
sns.lmplot(x = 'total_bill' , y= 'tip' , data = df , hue = 'sex', markers = ['o','v'])
```
